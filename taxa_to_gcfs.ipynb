```python
def mergeDictionary(dict_1, dict_2):
    dict_3 = {**dict_1, **dict_2}
    for key, value in dict_3.items():
        if key in dict_1 and key in dict_2:
            dict_3[key] =  dict_1[key] + dict_2[key]
    return  dict_3

mergeDictionary({"a": 1, "b": 2}, {"a": 6, "c": 3})
```

```python
prev_taxa = 'Species'
prev_dict = species_to_gcfs
wanted_cols = taxa.columns[1:7]
for col in wanted_cols[::-1]: # Genus --> Kingdom
     # print(col)
     # get this_taxa -> all prev_taxa in this group
     this_taxa = taxa.dropna().groupby(col)[prev_taxa].unique()
     # print(this_taxa)
     
     this_dict = {}
     for this_taxa_name, lower_taxa_list in this_taxa.items():
          merged_dict = {}
          for lower_taxa in lower_taxa_list:
               merged_dict = mergeDictionary(merged_dict, prev_dict[lower_taxa])
          this_dict[this_taxa_name] = merged_dict
          jsonified = []
          for gcf, hits in merged_dict.items():
               jsonified.append({"gcf": gcf, "hits": hits})
          with open('Data/Taxonomy/' + col + '/' + this_taxa_name + '.json', "w+") as taxa_f:
               json.dump(jsonified, taxa_f)
     
     prev_taxa = col
     prev_dict = this_dict

# genus = taxa.dropna().groupby('Genus')['Species'].unique()
# for index, value in genus.items():
#      print(index, value)
# prev_dict
```
