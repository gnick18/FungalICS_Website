```python
import json
import pandas as pd
```

```python
references = pd.read_csv('Data/ReferenceTable.tsv', sep='\t')
gcf_keys = pd.read_csv('Data/GCFs_Keys.tsv', sep='\t')
```

```python
merged = pd.merge(references, gcf_keys, 'left', left_on='BGC_Name', right_on='Cluster_Name')

merged
```

```python
cluster_to_accession = {}
for  index, row in merged.iterrows():
    cluster_to_accession[row['Cluster_Name']] = row['NCBI_Accession']

cluster_to_accession
```

```python
with open('Data/RefClusterToAccession.json', 'w+') as f:
    json.dump(cluster_to_accession, f)
```
