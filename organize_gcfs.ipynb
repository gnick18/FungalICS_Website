```python
import pandas as pd
import json
from collections import defaultdict

df = pd.read_table("Data/GCFs_Keys.tsv")
# print(df)
```

```python
gcf_keys_dict = defaultdict(list)
for _, row in df.iterrows():
    gcf_keys_dict[row['FinalGCFs']].append({
        "Cluster_Name": row['Cluster_Name'],
        "NCBI_Acession": row['NCBI_Accession']})

print(gcf_keys_dict)
```

```python
gcf_keys_json = json.dumps(gcf_keys_dict)

with open("Data/GCF_Keys.json", "w+") as f:
    f.write(gcf_keys_json)
```
